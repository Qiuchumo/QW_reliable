```python
np.set_printoptions(precision=0, suppress=True)
Savefig = 0 # 
plt.style.use('default')
elec_data = pd.read_csv('XZmulti_12.csv')

# 
companies_num = elec_data.counts.unique()
companies = len(companies_num)  # companies=7 7
company_lookup = dict(zip(companies_num, range(len(companies_num))))
company = elec_data['company_code'] = elec_data.counts.replace(company_lookup).values  # XZsingal
# companys = elec_data.counts.values - 1 # 

# 
company_ABC = elec_data.company.unique()
companiesABC = len(company_ABC)  # companies=7 7
company_lookup_ABC = dict(zip(company_ABC, range(len(company_ABC))))
companyABC = elec_data['company_ABC'] = elec_data.company.replace(company_lookup_ABC).values  # XZsingal
# companys = elec_data.counts.values - 1 # 
# elec_count = elec_data.counts.values
# print(companyABC)
# print(companiesABC)
# 
SNR = np.random.normal(0, 1, size=[len(elec_data.Year.values), 3])

# #
elec_tem = elec_data.Tem.values # x2
elec_tem1 = (elec_tem - np.mean(elec_tem)) / np.std(elec_tem)
elec_hPa = elec_data.hPa.values + SNR[:, 0]  # x3
elec_hPa1 = (elec_hPa - np.mean(elec_hPa)) / np.std(elec_hPa)
elec_RH = elec_data.RH.values + SNR[:, 1] # x3
elec_RH1 = (elec_RH - np.mean(elec_RH)) / np.std(elec_RH)
elec_Lux = elec_data.Lux.values + SNR[:, 2] # x3
elec_Lux1 = (elec_Lux - np.mean(elec_Lux)) / np.std(elec_Lux)


# 
elec_year = elec_data.Year.values  # x1
elec_year1 = (elec_year - np.mean(elec_year)) / np.std(elec_year)
data_cs_year = elec_year
# print(data_cs_year)

elec_Pca = np.vstack((elec_tem1, elec_hPa1, elec_RH1, elec_Lux1)).T   # 
# elec_Pca2 = np.vstack((elec_tem, elec_hPa, elec_RH, elec_Lux)).T   # 
# np.savetxt('XZ_nomean.csv', elec_Pca2, delimiter = ',')
# =============================================================================================
# # PCA
# x, z= pcaa(elec_Pca);  XX = np.array(x); ZZ = np.array(z)
# 4299%; 
# =============================================================================================
# #n_components
pca = PCA(n_components=2)
pca.fit(elec_Pca)
# Xtransform
elec_Pca1 = pca.transform(elec_Pca)
elec_Pca1 = np.array(elec_Pca1)

elec_Pca_char1 = elec_Pca1[:, 0] # 1
elec_Pca_char2 = elec_Pca1[:, 1] # 2
print('\n')

# elec_data.Fault.values[48] =2000
# /Y100100
elec_faults = 1000 * (elec_data.Fault.values / elec_data.Nums.values)  # ,
# elec_faults1 = (elec_faults - np.mean(elec_faults)) / np.std(elec_faults)

# 6,21*6
elec_faults2 = np.array([elec_faults[i*12:(i+1)*12] for i in np.arange(21)])
elec_year2 = np.array([elec_year[i*12:(i+1)*12] for i in np.arange(21)])
elec_char1 = np.array([elec_Pca_char1[i*12:(i+1)*12] for i in np.arange(21)])
elec_char2 = np.array([elec_Pca_char2[i*12:(i+1)*12] for i in np.arange(21)])
companyABC2 = np.array([companyABC[i*12:(i+1)*12] for i in np.arange(21)])

Plot_raw(elec_year2, elec_faults2, Savefig)
# 
# xs_char1 = shared(np.asarray(elec_Pca_char1))
# xs_char2 = shared(np.asarray(elec_Pca_char2))

# ys_faults = shared(np.asarray(elec_faults))
# xs_year = shared(np.asarray(data_cs_year))
# Num_shared = shared(np.asarray(companyABC))

def logit(x):
    return 1/(1+np.exp(-x))
def Invlogit(x):
    return tt.exp(x)/(1+tt.exp(x))
```

```python
# B
x_zhou = 12
Num = len(elec_faults[0:84])
knots = np.linspace(1, x_zhou, Num)

Num_5 = 5 * len(elec_faults[0:84])
model_knots = np.linspace(1, x_zhou, Num_5)

# @as_opx
basis_funcs = sp.interpolate.BSpline(knots, np.eye(Num_5), k=3) # eye()
Bx = basis_funcs(elec_year[0:84])  # x
# shared:symbolic variableashared variableaashared
Bx_ = shared(Bx)



Num = len(elec_faults[84:168])
knotsB = np.linspace(1, x_zhou, Num)

Num_5 = 5 * len(elec_faults[84:168])
model_knots1 = np.linspace(1, x_zhou, Num_5)

# @as_opx
basis_funcsB = sp.interpolate.BSpline(knotsB, np.eye(Num_5), k=3) # eye()
BxB = basis_funcsB(elec_year[84:168])  # x
# shared:symbolic variableashared variableaashared
Bx_B = shared(BxB)
```

```python
# 
# import ipyparallel as ipp
with pm.Model() as model_1:
    # define priors
    sigma = pm.HalfCauchy('sigma', 10)
    sigmaB = pm.HalfCauchy('sigmaB', 10)
    
    _a = pm.HalfCauchy('_a', 5.)
    _aB = pm.HalfCauchy('_aB', 5.)
    
    a0 = pm.Normal('a0', 0., 10.)
    a0B = pm.Normal('a0B', 0., 10.)
    
    _a = pm.Normal('_a', 0., 10., shape = (Num_5))
    _aB = pm.Normal('_aB', 0., 10., shape = (Num_5))
    
     = pm.Normal('', 0, sd=100) # 
    B = pm.Normal('B', 0, sd=100) # 
    
    theta1 = pm.Deterministic('theta1', a0 + (_a * _a).cumsum())
    theta1B = pm.Deterministic('theta1B', a0B + (_aB * _aB).cumsum())
    
    theta = pm.Deterministic('theta', tt.exp(Bx_.dot(theta1) + ))
    thetaB = pm.Deterministic('thetaB', tt.exp(Bx_B.dot(theta1B) + B))

    Observed = pm.Normal('Observed', mu=theta, sd=sigma, observed=elec_faults[0:84])  # 
    ObservedB = pm.Normal('ObservedB',mu=thetaB, sd=sigmaB, observed=elec_faults[84:168])  # 
#     step1 = pm.Slice([theta1, _a])
    start = pm.find_MAP()
    trace_1 = pm.sample(1000, start=start, njobs=1)
```

```python
pm.traceplot(trace_1)
plt.show()
```

```python
ax = pm.energyplot(trace_1)
bfmi = pm.bfmi(trace_1)
ax.set_title(f"BFMI = {bfmi:.2f}");
plt.show()
```

```python
# 
varnames2 = ['theta']
tmp = pm.df_summary(trace_1, varnames2)
betaMAP = tmp['mean'][np.arange(12)]
print(betaMAP)

with model_1:
    pp_trace = pm.sample_ppc(trace_1, 1000)
ip=0
fig, ax = plt.subplots(figsize=(8, 6))
x_plot = np.linspace(0.9, 12.1, 12)
low, high = np.percentile(pp_trace['Observed'], [25, 75], axis=0)

xp = elec_year2[ip*7:(ip+1)*7, :]   # 
yp = elec_faults2[ip*7:(ip+1)*7, :]
ax.plot(xp, yp, marker='o', alpha=.8)
ax.plot(x_plot, betaMAP[:], marker='*', alpha=.8)
ax.fill_between(x_plot, low[:12], high[:12],  alpha=0.5)
ax.plot(x_plot, pp_trace['Observed'].mean(axis=0)[:12], label="Spline estimate")

# ax.set_xlim(0, 1)
ax.legend()
plt.show()

ip=1
fig, ax = plt.subplots(figsize=(8, 6))
x_plot = np.linspace(0.9, 12.1, 12)
low, high = np.percentile(pp_trace['ObservedB'], [25, 75], axis=0)

xp = elec_year2[ip*7:(ip+1)*7, :]   # 
yp = elec_faults2[ip*7:(ip+1)*7, :]
ax.plot(xp, yp, marker='o', alpha=.8)
ax.plot(x_plot, betaMAP[:], marker='*', alpha=.8)
ax.fill_between(x_plot, low[:12], high[:12],  alpha=0.5)
ax.plot(x_plot, pp_trace['ObservedB'].mean(axis=0)[:12], label="Spline estimate")

# ax.set_xlim(0, 1)
ax.legend()
plt.show()
```
