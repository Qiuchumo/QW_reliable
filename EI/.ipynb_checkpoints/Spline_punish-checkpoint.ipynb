```python
# B
x_zhou = 12 #12
Num = len(elec_faults[:58])
knots = np.linspace(0, x_zhou, Num)

Num_5 = 5 * len(elec_faults[:58])#290
# Num_5 = shared(Num_51)
model_knots = np.linspace(1, x_zhou, Num_5)


# @as_opx
basis_funcs = sp.interpolate.BSpline(knots, np.eye(Num_5), k=3) # eye()
Bx = basis_funcs(xs_yearA.get_value())  # x
BxB = basis_funcs(xs_yearB.get_value())  # x
BxC = basis_funcs(xs_yearC.get_value())  # x
# shared:symbolic variableashared variableaashared
Bx_ = shared(Bx)
Bx_B = shared(BxB)
Bx_C = shared(BxC)
```

```python
Bx_.set_value(basis_funcs(xs_yearA.get_value()))
# 
with pm.Model() as model_1:
    # define priors
    alpha = pm.HalfCauchy('alpha', 10., testval=1.15)
    
    _a = pm.HalfCauchy('_a', 5.)
    a0 = pm.Normal('a0', 0., 10.)
    _a = pm.Normal('_a', 0., 20., shape = (Num_5))
    
    sigmaA = pm.Uniform('sigmaA', lower=0.01,upper=3 )
#      = pm.Normal('', 0, sd=sigmaA) # 
 
    theta1 = pm.Deterministic('theta1', a0 + (_a * _a).cumsum())
#     theta = pm.Deterministic('theta',   tt.exp(Bx_.dot(theta1)   +   + beta*elec_tem1[0:84]+ beta1*elec_RH1[0:84]))
    theta = pm.Deterministic('theta',  tt.exp( Bx_.dot(theta1) +  sigmaA*_a*_a  ))

    
    Observed = pm.Weibull('Observed',alpha=alpha, beta =theta, observed=ys_faultsA)  # 
#     ObservedB = pm.Weibull('ObservedB',alpha=alpha, beta =thetaB, observed=ys_faultsB)  # 
#     ObservedC = pm.Weibull('ObservedC',alpha=alpha, beta =thetaC, observed=ys_faultsC)  #   
#     step1 = pm.Slice([_a])
    start = pm.find_MAP()
    trace_1 = pm.sample(1000,start=start,  njobs=1)
```
